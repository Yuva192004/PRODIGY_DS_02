```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset 2 analysis
plt.figure(figsize=(8, 6))
sns.histplot(df2['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age of df2')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df1['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age of df1')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df2, hue='Survived', palette='viridis')
plt.title('Relationship between Age and Fare in df2')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
sns.pairplot(df2[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='husl')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=df2, palette='coolwarm')
plt.title('Survival Count by Sex in df2')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of df2')
plt.show()
```

```python
survived_count = df['Survived'].value_counts()
total_passengers = df.shape[0]
survived_proportion = survived_count[1] / total_passengers
not_survived_proportion = survived_count[0] / total_passengers
print("Proportion of passengers who survived:", survived_proportion)
print("Proportion of passengers who did not survive:", not_survived_proportion)
```
